<span style="font-size:30px;">   UNDERSAMPLING </span>

```python
#Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions
#Number of Fraudulent Transactions --> 492
legit_sample = legit.sample(n=492) # it will randobly select 492 rows in not frard
```

<span style="font-size:30px;">   Concatenating two DataFrames </span>

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0) # we want to add row wise of axis=0
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

<span style="font-size:30px;">   Splitting the data into Features & Targets (undersample)</span>

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
print("X =  \n"  ,X)
print("\n\n\n\n\n")
print("Y =  \n" ,Y)
```

<span style="font-size:30px;"> Split the data into Training data & Testing Data </span> 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
print("X = " ,X.shape,"X_train = ", X_train.shape, " X_test =" ,X_test.shape)
print("Y = " ,Y.shape,"Y_train = ", Y_train.shape, " Y_test =" ,Y_test.shape) 
```

<span style="font-size:30px;"> Building the Random Forest Classifier (RANDOM FOREST) - UnderSampaling
</span> 

```python
from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc = RandomForestClassifier(n_estimators = 100)
rfc.fit(X_train,Y_train)
# predictions
y_pred_rand_under = rfc.predict(X_test)
print(X_train.shape ,Y_train.shape, y_pred_rand_under.shape,Y_test.shape)
```

<span style="font-size:30px;">   Splitting the data into Features & Targets (Oversample)</span>

```python
from sklearn.model_selection import train_test_split as holdout
card_data = pd.read_csv('../input/creditcardfraud/creditcard.csv')
x_over = np.array(card_data.iloc[:, card_data.columns != 'Class'])
y_over = np.array(card_data.iloc[:, card_data.columns == 'Class'])
x_train_over, x_test_over, y_train_over, y_test_over = holdout(x_over, y_over, test_size=0.2, random_state=0)
print(y_test_over.shape,x_train_over.shape)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=2)
x_train_s, y_train_s = sm.fit_resample(x_train_over, y_train_over.ravel())
```

<span style="font-size:30px;"> Building the Random Forest Classifier (RANDOM FOREST) - OverSampaling
</span> 

```python
from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc1 = RandomForestClassifier() #object of random forest get created 
rfc1.fit(x_train_s, y_train_s) # we a random model on X train and Y train 
# predictions
y_pred_rand_over = rfc1.predict(x_test_over)

```

<span style="font-size:30px;">  Evaluating the classifier - Random Forest</span>


```python
#Evaluating the classifier
#printing every score of the classifier
#scoring in any thing
from sklearn.metrics import classification_report, accuracy_score,precision_score,recall_score,f1_score,matthews_corrcoef
from sklearn.metrics import confusion_matrix,plot_confusion_matrix,classification_report
print("The model used is Random Forest classifier")
acc= accuracy_score(y_test_over,y_pred_rand_over)
print("The accuracy is  {}".format(acc))
prec= precision_score(y_test_over,y_pred_rand_over)
print("The precision is {}".format(prec))
rec= recall_score(y_test_over,y_pred_rand_over)
print("The recall is {}".format(rec))
f1= f1_score(y_test_over,y_pred_rand_over)
print("The F1-Score is {}".format(f1))
MCC=matthews_corrcoef(y_test_over,y_pred_rand_over)
print("The Matthews correlation coefficient is {}".format(MCC))


cnf_matrix = confusion_matrix(y_test_over, y_pred_rand_over)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu", fmt='g')
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')

print(classification_report(y_test_over, y_pred_rand_over))

```
